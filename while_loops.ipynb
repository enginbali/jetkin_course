### break:

````python
n = 5
while n > 0:
    n -= 1
    if n == 2:
        break
    print(n)
print('Loop ended.')
````

### Continue:

````python
n = 5
while n > 0:
    n -= 1
    if n == 2:
        continue
    print(n) # when n is 2, the continue statement causes termination of that iteration. Thus, 2 isnt printed.
print('Loop ended.')
````

### while... else...

````python
n = 5
while n > 0:
    n -= 1
    print(n)
else:
    print('Loop done.')
````

````python
n = 5
while n > 0:
    n -= 1
    print(n)
    if n == 2:
        break
else:
    print('Loop done.') # This loop is terminated prematurely with break, so the else clause isnt executed.
````

One common situation is if you are searching a list for a specific item. You can use break to exit the loop if the item is found, and the else clause can contain code that is meant to be executed if the item isnt found:

````python
a = ['bayern', 'barca', 'city'] #, 'demirspor'
s = 'demirspor'

a[0] == s
````

````python
a = ['bayern', 'barca', 'city', 'demirspor'] #
s = 'demirspor'

i = 0
while i < len(a):
    if a[i] == s:
        # Processing for item found
        print('demirspor has been found!')
        break
    i += 1
else:
    # Processing for item not found
    print(s, 'not found in list.')
````

### Nested While Loops:

````python
# if -- elif nested
````

````python
age = 85
gender = 'F'

if age < 18:
    if gender == 'M':
        print('son')
    else:
        print('daughter')
elif age >= 18 and age < 65:
    if gender == 'M':
        print('father')
    else:
        print('mother')
else:
    if gender == 'M':
        print('grandfather')
    else:
        print('grandmother')
````

````python
# while can be nested also
````

````python
a = ['foo', 'bar']

while len(a):
    
    print(a.pop(0))
    b = ['baz', 'qux']
    
    while len(b):
        print('>', b.pop(0))
````

A break or continue statement found within nested loops applies to the nearest enclosing loop:

```python
while <expr1>:
    statement
    statement

    while <expr2>:
        statement
        statement
        break  # Applies to while <expr2>: loop

    break  # Applies to while <expr1>: loop
```

Additionally, while loops can be nested inside if/elif/else statements, and vice versa:

```python 
if <expr>:
    statement
    while <expr>:
        statement
        statement
else:
    while <expr>:
        statement
        statement
    statement
```

```python
while <expr>:
    if <expr>:
        statement
    elif <expr>:
        statement
    else:
        statement

    if <expr>:
        statement
```
